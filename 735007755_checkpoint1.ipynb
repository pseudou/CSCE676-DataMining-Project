## Part (A): Identification of Candidate Datasets

The three candidate datasets below are unified by a single domain — **online retail and e-commerce**. This connection is deliberate: it might allow cross-dataset insights (e.g., comparing co-purchase graph edges with association rules mined from transaction baskets, or linking review sentiment to product centrality in the graph). ### Dataset 1: Online Retail II (UCI Machine Learning Repository)

**Dataset name and source**  
Online Retail II, UCI Machine Learning Repository (Dataset ID: 502), donated by Dr. Daqing Chen.
https://archive.ics.uci.edu/dataset/502/online+retail+ii

**Why this dataset was selected**  
This dataset represents real transaction‑level retail data and is a natural fit for pattern mining tasks. Each invoice corresponds to a shopping basket, making the data well‑suited for frequent itemset mining and association rule discovery. Unlike synthetic datasets, it contains missing values, cancellations, and negative quantities, which introduces realistic data cleaning and modeling challenges. The presence of timestamps and customer identifiers also enables temporal analysis beyond unordered basket mining.

**In‑course techniques supported**
- Frequent itemset mining and association rules
- Clustering via customer purchase behavior
- Anomaly detection using cancellations and negative quantities

**Beyond‑course techniques enabled**
- Shopping Trajectory Representation Learning — Learns dense vector representations of entire customer purchase sequences (not just individual baskets), capturing temporal shopping behavior. Based on the KDD 2024 paper "Shopping Trajectory Representation Learning with Pre-training for E-commerce Customer Understanding"
- Debiased Price Optimization via Double Machine Learning — Applies causal inference frameworks to estimate price elasticity from observational transaction data while correcting for confounding. Based on the KDD 2025 Causal ML Workshop paper "A Debiased Machine Learning Framework for Optimizing Price Elasticity Models".
- Sequential pattern mining using customer‑level purchase sequences
- Statistical validation of association rules using chi‑squared tests and Bonferroni correction

**Dataset size and structure**
- 1,067,371 transaction line items
- 43.5 MB Excel file
- Variable‑length baskets grouped by invoice
- Time span: December 2009 – December 2011

**Data types**
- Invoice ID
- Product ID (StockCode)
- Quantity
- Price
- Timestamp
- Customer ID
- Country

**Target variable(s)**
- None (primarily unsupervised)
- Derived targets such as revenue and cancellation indicators

**Licensing or usage constraints**
- CC BY 4.0 — free for academic use with attribution

Amazon Product Co‑Purchasing Network, Stanford Network Analysis Project (SNAP).

This dataset captures product‑level relationships through co‑purchase behavior, forming a large but tractable graph. - Graph neural networks for link prediction
- Node embedding methods such as node2vec and DeepWalk

**Dataset size and structure**
- 334,863 nodes (products)
- 925,872 undirected edges
- Edge list size approximately 16 MB
- Single connected component

**Data types**
- Node IDs
- Edge pairs representing co‑purchase relationships
- Community membership lists

**Target variable(s)**
- Ground‑truth community labels
- Link prediction targets

## Part B — Comparative Analysis of Candidate Datasets

The three candidate datasets are compared below across data mining capabilities, data quality, feasibility of algorithmic analysis, and broader bias and ethical considerations. The goal of this comparison is to understand the trade‑offs involved in each dataset and to assess how well each supports both course techniques and at least one additional method beyond the course.

| Dimension | Online Retail II (Transactions) | Amazon Co-Purchase Graph | Amazon Musical Instruments Reviews |
|---|---|---|---|
| Supported data mining tasks | Frequent itemsets + association rules (course); clustering + anomaly detection (course); sequential patterns (beyond) | Centrality/PageRank (course); community detection (course); node embeddings + GNN link prediction (beyond) | TF-IDF/embeddings (course); clustering (course); topic modeling + transformer embeddings (beyond) |
| Data quality issues | Missing Customer_ID; cancellations/returns; duplicates; long-tail sparsity | No edge weights; edges lack frequency; overlapping communities | Noisy text; self-selection; missing helpfulness votes |
| Algorithmic feasibility | FP-Growth preferred; Apriori needs high support; sequences need filtering | Manageable on single machine with efficient libs | Feasible on 5-core subset; transformers need batching |
| Bias considerations | Wholesale/seasonality effects; promotions | Recommendation/merchandising effects; popularity dominance | Reviewer self-selection; extreme ratings overrepresented |
| Ethical considerations | Anonymized transactions; low individual risk | Product-level only; low individual risk | Public reviews; analyze aggregates, not individuals |

### Summary of Trade‑offs

- The **transaction dataset** offers strong support for pattern mining and temporal analysis but requires careful handling of noise and sparsity.
- The **graph dataset** enables rigorous evaluation of graph algorithms due to its ground‑truth communities but abstracts away transaction frequency and timing.
- The **review dataset** provides rich textual information suitable for semantic analysis but introduces subjectivity and self‑selection bias.

## Part C — Dataset Selection

### Selected dataset
Online Retail II (UCI Machine Learning Repository)

---

### Reasons for selection

The Online Retail II dataset was selected because it provides a clean and realistic foundation for studying transaction‑level purchasing behavior while still allowing extensions beyond basic pattern mining.

First, the structure of the data directly supports frequent itemset mining and association rule analysis. Each invoice corresponds to a shopping basket containing multiple products, which makes the mapping from raw data to itemsets straightforward. This allows the analysis to focus on algorithmic behavior and pattern quality rather than on artificial data transformations.

Second, the dataset includes timestamps and customer identifiers, which makes it possible to move beyond unordered baskets and analyze purchase behavior over time. By grouping transactions by customer and ordering them temporally, the data can be transformed into sequences. This enables sequential pattern mining and allows a direct comparison between patterns discovered with and without temporal information.

Third, the dataset reflects real operational data rather than a curated benchmark. Missing customer identifiers, cancellation invoices, and negative quantities introduce realistic noise and edge cases. Handling these issues requires explicit assumptions and preprocessing decisions, which helps ground the analysis in practical data mining considerations rather than idealized inputs.

Fourth, the dataset size strikes a good balance between realism and feasibility. With just over one million transaction rows, it is large enough to exhibit sparsity and long‑tail effects, but still small enough to be processed on a single machine with careful filtering and threshold selection. This makes iterative experimentation practical without requiring distributed infrastructure.

Finally, the dataset leaves room for multiple analytical directions if needed. In addition to pattern mining, it supports customer‑level clustering and anomaly detection based on returns and cancellations, providing flexibility as the project evolves.

---

### Trade‑offs and limitations

The dataset does not contain rich textual data suitable for deep natural language analysis. Product descriptions are short and inconsistent, which limits the usefulness of advanced text modeling techniques.

The data is primarily unsupervised. While derived targets such as revenue or cancellation indicators can be constructed, there is no natural labeled outcome variable for standard supervised learning tasks.

Sequential analysis depends on the availability of customer identifiers. Transactions without CustomerID must be excluded from sequence construction, which reduces coverage and may introduce selection bias toward repeat customers.

### Cleaning decisions and rationale

The dataset is transactional line-items. Cleaning focuses on removing rows that cannot represent a valid purchase event while preserving enough data for both basket mining and sequence mining.

### Standardize columns and types

```python
df = raw.copy()

df.columns = (
    df.columns.str.strip()
              .str.replace(r"\s+", "_", regex=True)
)

# Common column names in this dataset
df["Invoice"] = df["Invoice"].astype(str).str.strip()
df["StockCode"] = df["StockCode"].astype(str).str.strip()
df["Description"] = df["Description"].where(df["Description"].notna(), np.nan)
df.loc[df["Description"].notna(), "Description"] = df.loc[df["Description"].notna(), "Description"].str.strip()
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["Customer_ID"] = pd.to_numeric(df.get("Customer_ID"), errors="coerce")
df["Country"] = df["Country"].astype(str).str.strip()
```

### Identify cancellations and returns
Invoices starting with C are cancellations. Negative quantities also indicate returns/adjustments.

```python
df["IsCancellationInvoice"] = df["Invoice"].str.startswith("C", na=False)
df["IsReturnLine"] = df["Quantity"] < 0
```

### Apply row-level validity filters
Assumptions:

Keep only rows with valid timestamps and product codes.

For basket mining, keep purchases only: Quantity > 0 and Price > 0.

Keep missing Customer_ID for basket-level EDA; filter later for sequence mining.

```python
before = len(df)

df = df.dropna(subset=["InvoiceDate", "Invoice", "StockCode"])
df = df[df["Quantity"].notna() & df["Price"].notna()]

# Keep only positive purchase lines
df = df[(df["Quantity"] > 0) & (df["Price"] > 0)]

after = len(df)
before, after, (before - after)
```

### Create derived fields used throughout EDA

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
df["Date"] = df["InvoiceDate"].dt.date
df["YearMonth"] = df["InvoiceDate"].dt.to_period("M").astype(str)
df["Hour"] = df["InvoiceDate"].dt.hour
df["Weekday"] = df["InvoiceDate"].dt.day_name()
df.sample(10)
```

## Co-occurrence sparsity (how dense is the item graph implied by baskets?)

A practical way to quantify sparsity is to:

1. restrict to a frequent-item subset (otherwise the pair space is enormous),

2. compute how many item pairs actually occur vs. how many are possible.

### Build co-occurrence pairs for top-N items

```python
from itertools import combinations
from collections import Counter

N = 200  # adjust based on runtime
top_items = set(prod_inv_freq.head(N).index)

pairs = Counter()
inv_groups = df[df["StockCode"].isin(top_items)].groupby("Invoice")["StockCode"].unique()

for items in inv_groups:
    items = np.unique(items)
    if len(items) >= 2:
        for a, b in combinations(sorted(items), 2):
            pairs[(a, b)] += 1

n_possible = N * (N - 1) // 2
n_observed = len(pairs)
density = n_observed / n_possible

{"topN": N, "possible_pairs": n_possible, "observed_pairs": n_observed, "pair_density": density}
```

```python
top_pairs = pd.DataFrame(
    [(a, b, c) for (a, b), c in pairs.most_common(20)],
    columns=["item_a", "item_b", "cooccur_invoices"]
)
top_pairs
```

## Part E — Initial Insights and Direction

This section summarizes the key observations from the exploratory data analysis and outlines how these findings motivate the analytical direction of the project.

---

### Key Observations from EDA

- **Strong long‑tail behavior in product frequency.**  
  Over 20% of products appear in fewer than 10 invoices, while a small number of products appear in thousands of invoices. This indicates that global frequency‑based thresholds will disproportionately favor a small subset of popular items.

- **Revenue is highly concentrated among a small fraction of products.**  
  The top 1% of products account for over 20% of total revenue, and the top 10% account for more than 60%. This suggests that support‑based pattern mining may not align with revenue impact, motivating alternative evaluation criteria.

- **Basket sizes are highly skewed with extreme outliers.**  
  Most invoices contain fewer than 30 distinct items, but a small number of invoices contain several hundred items. These large baskets are likely wholesale orders and can dominate co‑occurrence statistics if not handled carefully.

- **Customer identifiers are sparsely populated and unevenly distributed.**  
  Only a subset of invoices contain valid Customer_ID values. Customer‑level and sequential analyses therefore apply to a filtered population of repeat, identifiable customers rather than the full transaction set.

- **Temporal structure is non‑uniform.**  
  Purchases exhibit strong monthly seasonality and clear intraday patterns. Time gaps between purchases vary widely across customers, indicating that temporal ordering may reveal structure not captured by unordered baskets.

- **Item co‑occurrence is sparse even among frequent products.**  
  Pairwise co‑occurrence density remains low after restricting to the most frequent items, suggesting that meaningful associations are localized rather than global.

---

### Implications for Analysis

These observations indicate that treating transactions as unordered baskets captures only part of the underlying structure. High‑support itemsets are dominated by popular products and large baskets, while many products participate in meaningful but infrequent or temporally ordered patterns.

---

### Initial Hypotheses

- High support thresholds in association rule mining will miss meaningful patterns involving less frequent products.
- Sequential patterns derived from customer‑level purchase histories will reveal structure that is not visible in unordered basket analysis.
- Filtering or stratifying by basket size, customer activity level, or time period will materially affect the quality and interpretability of discovered patterns.

---

### Preliminary Research Questions

- How do frequent itemsets differ from sequential patterns when evaluated on the same subset of customers?
- What types of product relationships emerge only when temporal order is considered?
- How sensitive are discovered patterns to basket outliers and customer filtering choices?

---

### Direction Going Forward

Based on the EDA, the next phase of the project will focus on:
1. Mining frequent itemsets and association rules with carefully chosen support thresholds and basket filtering.
2. Constructing customer‑level purchase sequences and applying sequential pattern mining.
3. Comparing unordered and temporal patterns to understand what information is gained by incorporating time.
